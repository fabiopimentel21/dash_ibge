```python
duplicadas = df_demogarfico_01_editada.columns[df_demogarfico_01_editada.columns.duplicated()]
print(duplicadas)
```

```python
def renomear_colunas_duplicadas(colunas):
    contagem = {}
    novas_colunas = []
    for col in colunas:
        if col in contagem:
            contagem[col] += 1
            novas_colunas.append(f"{col}_{contagem[col]}")
        else:
            contagem[col] = 1
            novas_colunas.append(col)
    return novas_colunas

df_demogarfico_01_editada.columns = renomear_colunas_duplicadas(df_demogarfico_01_editada.columns)
```

```python
df_demogarfico_01_editadaa = pd.DataFrame(df_demogarfico_01_editada)

plt.figure(figsize=(8,5))
plt.bar(df_demogarfico_01_editada['NM_BAIRRO'], df_demogarfico_01_editada['Domicílios Particulares Permanentes Ocupados, Com 5 moradores no domicílio'], color='steelblue')
plt.xticks(rotation=90)
plt.title('Qtd')
plt.xlabel('bairro')
plt.ylabel('qtd moradores')
plt.tight_layout()
plt.show
```
